```python
import pandas as pd
import sqlalchemy as sa
import plotly.express as px

# Conectar a Northwind
engine = sa.create_engine('sqlite:///northwind.db')

# Cargar tabla de clientes
df_customers = pd.read_sql('SELECT * FROM Customers', engine)
df_customers.head()
```

```python
# Cargar ventas con detalle
df_orders = pd.read_sql('''
    SELECT 
        o.OrderID,
        o.OrderDate,
        c.Country,
        p.ProductName,
        cat.CategoryName,
        od.Quantity,
        od.UnitPrice,
        od.Quantity * od.UnitPrice AS TotalVenta
    FROM Orders o
    JOIN Customers c ON o.CustomerID = c.CustomerID
    JOIN [Order Details] od ON o.OrderID = od.OrderID
    JOIN Products p ON od.ProductID = p.ProductID
    JOIN Categories cat ON p.CategoryID = cat.CategoryID
''', engine)

df_orders.head()
```

```python
# Exportar a CSV para Power BI
df_orders.to_csv('ventas_northwind.csv', index=False)
print(f"Exportado: {len(df_orders)} filas")
```
